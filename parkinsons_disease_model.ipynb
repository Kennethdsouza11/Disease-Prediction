```python
import numpy as np 
import pandas as pd 
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
```

```python
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
from pandas.core.dtypes.common import is_numeric_dtype
for column in parkinson_data.columns:
    if is_numeric_dtype(parkinson_data[column]):
        continue
    else:
        parkinson_data[column] = label.fit_transform(parkinson_data[column])
```

```python
#seperating the data and labels
X = parkinson_data.drop(columns = ['status', 'name'], axis = 1)
Y = parkinson_data['status']
```

```python
print(X)
```

```python
print(Y)
```

```python
parkinson_data.isnull().sum()
```

```python
#Data Standardisation
scaler = StandardScaler()
scaler.fit(X)
standardised_data = scaler.transform(X)
```

```python
X = standardised_data
```

```python
print(X)
```

```python
#Splitting the data into training and testing data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, stratify = Y, random_state= 1)
```

```python
print(X.shape, X_train.shape, X_test.shape, Y.shape, Y_train.shape, Y_test.shape)
```

```python
#Training the model
model = svm.SVC(kernel = 'linear')
model.fit(X_train, Y_train)
```

```python
Y_pred= model.predict(X_test)
Y_prediction_accuracy = accuracy_score(Y_test, Y_pred)
print("Model's accuracy score is (related to test data) : ", Y_prediction_accuracy)
```

```python
# Hyperparameter tuning using Grid Search with cross-validation
param_grid = {
    'C': [0.1, 1, 10, 100],
    'gamma': [1, 0.1, 0.01, 0.001],
    'kernel': ['rbf', 'poly', 'sigmoid']
}
grid = GridSearchCV(SVC(), param_grid, refit=True, verbose=2, cv=5)
grid.fit(X_train, Y_train)
best_svm_model = grid.best_estimator_
best_svm_model.fit(X_train,Y_train)
```

```python
Y_pred_tuned = best_svm_model.predict(X_test)
Y_pred_tuned_accuracy = accuracy_score(Y_test, Y_pred_tuned)
print("Model's accuracy score is (related to training data) : ", Y_pred_tuned_accuracy)
```

```python
from sklearn.metrics import r2_score
score_train = r2_score(Y_test, Y_pred_tuned)
print("The model's goodness based on the training data according to r2_score is : ", score_train)
```

```python
from sklearn.metrics import confusion_matrix
cf_matrix = confusion_matrix(Y_test, Y_pred_tuned)
import seaborn as sns
sns.heatmap(cf_matrix, annot = True)
```
